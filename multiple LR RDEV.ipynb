## splitting out data into x and y values 

```python
x = dev.iloc[:,:4]
y = dev.iloc[:,4:]
```

## establishing our train and test size 

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=0)
```

## applying the linear regression 

```python
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

## predicting using linear regression

```python
y_pred_train = regressor.predict(X_train) 
y_pred_test = regressor.predict(X_test)
```

## PLOTTING THE PREDICTED VALUES ON THE ORIGINAL VALUES USING THE TEST DATA

```python
for column in X_test.columns: 
    plt.scatter(X_test[column],y_test,label= column)
    plt.plot(X_test[column],y_pred_test,label= column)
plt.show()
```

## finding the coefficient


```python
print(f'coefficient of the first linear regression: {regressor.coef_}')
print(f'intercept of the first linear regression: {regressor.intercept_}')
```
